### Реализуем градиентный спуск для задачи поиска оптимальных коэффициентов в MSE регрессии!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Имеем 1000 объектов и 10 признаков у каждого (+таргет)!

Обучим модель линейной регрессии:

$$
a(x) = \beta_1 d_{1} + \beta_2 d_{2} + \beta_3 d_{3} + \beta_4 d_{4} + \beta_5 d_{5} + \beta_6 d_{6} + \beta_7 d_{7} + \beta_8 d_{8} + \beta_9 d_{9} + \beta_{10} d_{10} + \beta_0
$$

Которая минимизирует MSE:

$$
Q(a(X), Y) = \sum_i^{1000} (a(x_i) - y_i)^2
$$

```python
data = pd.read_csv('data.csv')
data.head(6)
```

Обучим коэффициенты линейной регрессии с помощью библиотеки <a href="https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html"> **sklearn** </a>

Отдельно выведем оценку свободного коэффициента  ($\beta_0$ при $d_0 = 1$)

```python
from sklearn.linear_model import LinearRegression
### Your code is here
X = data.drop(['target'], axis = 1)
y = data['target']

linear_model = LinearRegression()
linear_model.fit(X, y)
print(linear_model.coef_)
print(linear_model.intercept_)
```

```python
class GradientDescentMse:
    """
    Базовый класс для реализации градиентного спуска в задаче линейной МНК регрессии 
    """
    def __init__(self, samples: pd.DataFrame, targets: pd.DataFrame,
                 learning_rate: float = 1e-3, threshold = 1e-6, copy: bool = True):
        self.samples = samples
        self.targets = targets
        self.learning_rate = learning_rate
        self.threshold = threshold
        self.iteration_loss_dict = {}
        self.beta = np.ones(self.samples.shape[1])
        self.iteration_num = 0
        
    def add_constant_feature(self):
        if 'const' not in self.samples.columns:
            self.samples['const'] = 1
            self.beta = np.append(self.beta, 1)
        
    def calculate_mse_loss(self) -> float:
        loss = (self.samples.values @ self.beta) - self.targets.values
        mse = np.mean(loss**2)
        return mse
        """
        Метод для расчета среднеквадратической ошибки
        :return: среднеквадратическая ошибка при текущих весах модели : float
        """
    def calculate_gradient(self) -> np.ndarray:
        N = self.samples.shape[0]
        predictions = self.samples.values @ self.beta
        errors = predictions - self.targets.values
        gradient = (2 / N) * (self.samples.values.T @ errors)
        
        return gradient
        """
        Подсчет значений вектора-градиента, матричное умножение (np.array или DataFrame), вычисление производной в матричной форме
        """
    def iteration(self): 
        """
        Обновляем веса модели в соответствии с текущим вектором-градиентом
        """
        ### Your code is here
        gradient = self.calculate_gradient()
        next_beta = self.beta - self.learning_rate * gradient
        return next_beta
        
    def learn(self):
        """
        Итеративное обучение весов модели до срабатывания критерия останова
        """
        current_mse = self.calculate_mse_loss() 
        self.iteration_loss_dict[0] = current_mse
        
        while True:
            next_beta = self.iteration() 
            beta_change = np.max(np.abs(next_beta - self.beta))
            self.beta = next_beta
            self.iteration_num += 1
            current_mse = self.calculate_mse_loss() 
    
            self.iteration_loss_dict[self.iteration_num] = current_mse
            if beta_change < self.threshold:
                break
```

```python
GD = GradientDescentMse(samples=X, targets=y)
GD.add_constant_feature()
GD.learn()
```

```python
print('Веса модели при переменных d1, d2, ..., d10 равны соответственно: \n\n' + str(GD.beta))
```

## Исследуем траектории обучения (learning paths)
На каждом графике отображено итеративное изменение значений минимизируемого функционала в зависимости от значений threshold и learning rate. 
В виду того, что мы наблюдаем монотонное уменьшение ошибки на каждой итерации, наилучший результат достигается при комбинации значений learning rate = [0.1, 0.5] и threshold = [1e-04, 1e-05] 

```python
# Изменение базовых настроек matplotlib

plt.rcParams.update({     
    'figure.dpi': 200,              
    'axes.titlesize': 10,          # Размер заголовка
    #'legend.fontsize': 10,        # Размер легенды
    'grid.alpha': 0.3,              
    'lines.linewidth': 2,           
    'axes.grid': True,              
    'font.family': 'DejaVu Sans',
    
    #'xtick.labelsize': 14,
    'xtick.major.size' : 15,
    'xtick.major.width' : 3,
    'xtick.labelsize' : 15,
    'xtick.color' : '#FF5533',
    #'ytick.labelsize': 14
    'ytick.major.size' : 15,
    'ytick.major.width' : 3,
    'ytick.labelsize' : 15,
    'ytick.color' : '#FF5533'
})
```

```python
fig = plt.figure()
fig.set_size_inches(13, 10)
rates = [0.1, 0.05, 0.01, 0.005, 0.001]
thresholds = [0.01, 0.001, 0.0001, 0.00001]

for i in range(len(thresholds)):
    thresh = thresholds[i]
    ax = fig.add_subplot(2, 2, i+1)
    Q_values = []  
    for lr in rates:
        GD = GradientDescentMse(samples=X, targets=y, 
                        learning_rate=lr, threshold=thresh)
        GD.add_constant_feature()
        GD.learn()
        learning_path = GD.iteration_loss_dict   
        last_mse = list(learning_path.values())[-1] 
        ax.plot(list(learning_path.keys()), list(learning_path.values()))    
        Q_values.append(str(round(last_mse, ndigits=4)))
    ax.set_ylim(0, 100)
    ax.set_xlim(0, 2000)
    ax.set_ylabel('Среднеквадратическая ошибка')
    ax.set_xlabel('Номер итерации')
    ax.legend([f'At learning rate {rates[j]} and threshold {thresh}' + ' last MSE = ' + Q_values[j] 
               for j in range(len(rates))])

fig.tight_layout() 
plt.show()
```
